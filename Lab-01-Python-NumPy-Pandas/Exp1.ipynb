# Experiment 1: Create NumPy Array from Python List

**Objective:** Create a NumPy array from a Python list, display it, and print its shape and data type.

```python
import numpy as np

# Create a Python list
python_list = [10, 20, 30, 40, 50]

# Convert to NumPy array
arr = np.array(python_list)

# Display the array
print('NumPy Array:')print(arr)

# Print shape
print('\nShape:', arr.shape)

# Print data type
print('Data Type:', arr.dtype)

# Print number of dimensions
print('Number of Dimensions:', arr.ndim)
```
