# Experiment 7: EDA - describe() and dtypes

**Objective:** Understand data types and statistical summary.

```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'ID': [1, 2, 3, 4, 5],
    'Score': [85.5, 90.2, 78.9, 92.1, 88.7],
    'Attempts': [1, 2, 1, 3, 2],
    'Grade': ['A', 'A+', 'B+', 'A+', 'A']
}

df = pd.DataFrame(data)

# Data types of each column
print('Data Types:')
print(df.dtypes)

print('\n' + '='*50)

# Descriptive statistics
print('\nDescriptive Statistics:')
print(df.describe())

print('\n' + '='*50)

# Get statistics for specific columns
print('\nStatistics for Score column:')
print(df['Score'].describe())

# Get individual statistics
print('\nIndividual Statistics:')
print('Mean:', df['Score'].mean())
print('Median:', df['Score'].median())
print('Std Dev:', df['Score'].std())
print('Min:', df['Score'].min())
print('Max:', df['Score'].max())
```
