# Experiment 4: Pandas DataFrame Creation

**Objective:** Create and manipulate Pandas DataFrames.

```python
import pandas as pd
import numpy as np

# Create a DataFrame from dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'Score': [85.5, 90.2, 78.9, 92.1, 88.7],
    'City': ['NYC', 'LA', 'Chicago', 'Boston', 'Seattle']
}

df = pd.DataFrame(data)
print('DataFrame:')
print(df)

# DataFrame properties
print('\nShape:', df.shape)
print('Columns:', df.columns.tolist())
print('Index:', df.index.tolist())
print('Data types:\n', df.dtypes)

# Accessing columns and rows
print('\nAge column:')
print(df['Age'])

print('\nFirst row:')
print(df.iloc[0])

# Basic statistics
print('\nBasic statistics:')
print(df.describe())
```
