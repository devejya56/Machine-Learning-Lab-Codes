# Experiment 3: Create Random Arrays

**Objective:** Generate random arrays using various NumPy functions.

```python
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Random floats between 0 and 1
print('Random floats (0-1):')
random_floats = np.random.rand(3, 3)
print(random_floats)

# Random integers between 1 and 100
print('\nRandom integers (1-100):')
random_ints = np.random.randint(1, 100, 10)
print(random_ints)

# Random normal distribution
print('\nRandom normal distribution:')
random_normal = np.random.randn(5)
print(random_normal)

# Random numbers from uniform distribution
print('\nRandom uniform (10-20):')
random_uniform = np.random.uniform(10, 20, 5)
print(random_uniform)

# Random choice
print('\nRandom choice from list:')
choices = np.random.choice([1, 2, 3, 4, 5], 5)
print(choices)
```
