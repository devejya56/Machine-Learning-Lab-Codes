# Experiment 6: EDA - head(), tail(), info()

**Objective:** Perform exploratory data analysis using head, tail, and info methods.

```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'ID': range(1, 11),
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Henry', 'Iris', 'Jack'],
    'Age': [25, 30, 35, 28, 32, 27, 29, 31, 26, 33],
    'Salary': [50000, 60000, 75000, 62000, 68000, 55000, 65000, 70000, 58000, 72000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'Sales', 'HR', 'Finance', 'Sales', 'IT', 'Finance']
}

df = pd.DataFrame(data)

# Display first n rows
print('First 3 rows:')
print(df.head(3))

# Display last n rows
print('\nLast 3 rows:')
print(df.tail(3))

# Info about the dataframe
print('\nDataFrame Info:')
df.info()

# Number of rows and columns
print('\nShape:', df.shape)
print('Total elements:', df.size)
```
