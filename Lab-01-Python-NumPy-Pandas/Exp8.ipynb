# Experiment 8: Data Type Conversions

**Objective:** Convert between different data types in Pandas.

```python
import pandas as pd
import numpy as np

# Create a DataFrame with mixed data types
data = {
    'ID': ['1', '2', '3', '4', '5'],  # Strings
    'Score': ['85.5', '90.2', '78.9', '92.1', '88.7'],  # String floats
    'Active': ['Yes', 'No', 'Yes', 'Yes', 'No']  # String booleans
}

df = pd.DataFrame(data)

print('Original Data Types:')
print(df.dtypes)
print('\nDataFrame:')
print(df)

# Convert string to integer
df['ID'] = df['ID'].astype(int)

# Convert string to float
df['Score'] = df['Score'].astype(float)

# Convert string to boolean
df['Active'] = df['Active'] == 'Yes'

print('\nAfter Conversion:')
print('Data Types:')
print(df.dtypes)
print('\nDataFrame:')
print(df)

# Using pd.to_numeric for more control
test_data = pd.Series(['1', '2', 'three', '4'])
print('\nUsing to_numeric (with errors=coerce):')
converted = pd.to_numeric(test_data, errors='coerce')
print(converted)
```
