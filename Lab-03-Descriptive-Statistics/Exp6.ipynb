# Lab 03: Exp6 - Covariance Matrix

```python
import pandas as pd
import numpy as np

data = {'math': [72, 68, 75, 80, 85], 'reading': [70, 75, 80, 85, 88], 'writing': [75, 72, 78, 82, 86]}
df = pd.DataFrame(data)

print('Covariance Matrix:')
print(df.cov())
print('\nCovariance (math vs reading):', np.cov(df['math'], df['reading'])[0,1])
```
