# Lab 03: Exp5 - Correlation Matrix

```python
import pandas as pd
import numpy as np

data = {'math': [72, 68, 75, 80, 85], 'reading': [70, 75, 80, 85, 88], 'writing': [75, 72, 78, 82, 86]}
df = pd.DataFrame(data)

print('Correlation Matrix:')
print(df.corr())
print('\nPearson Correlation (math vs reading):', df['math'].corr(df['reading']))
```
