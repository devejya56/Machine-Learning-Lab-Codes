# Lab 03: Exp8 - Boxplots for Outlier Detection

```python
import pandas as pd
import numpy as np

data = {'scores': [72, 68, 75, 80, 85, 90, 78, 82, 88, 92, 120]}
df = pd.DataFrame(data)

Q1 = df['scores'].quantile(0.25)
Q3 = df['scores'].quantile(0.75)
IQR = Q3 - Q1

print('Q1:', Q1)
print('Q3:', Q3)
print('IQR:', IQR)
print('Lower Bound:', Q1 - 1.5*IQR)
print('Upper Bound:', Q3 + 1.5*IQR)

outliers = df[(df['scores'] < Q1 - 1.5*IQR) | (df['scores'] > Q3 + 1.5*IQR)]
print('\nOutliers detected:')
print(outliers)
```
