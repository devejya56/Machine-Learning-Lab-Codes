# Lab 03: Exp3 - Variance and Standard Deviation

```python
import pandas as pd
import numpy as np

data = {'scores': [72, 68, 75, 80, 85, 90, 78, 82, 88, 92]}
df = pd.DataFrame(data)

print('Variance:', df['scores'].var())
print('Std Dev:', df['scores'].std())
print('Range:', df['scores'].max() - df['scores'].min())
```
