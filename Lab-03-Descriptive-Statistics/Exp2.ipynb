# Lab 03: Exp2 - Mean, Median, Mode

```python
import pandas as pd
from scipy import stats

data = {'scores': [72, 68, 75, 80, 85, 90, 78, 82, 88, 92]}
df = pd.DataFrame(data)

print('Mean:', df['scores'].mean())
print('Median:', df['scores'].median())
print('Mode:', stats.mode(df['scores'], keepdims=True).mode[0])
```
