# Lab 03: Exp4 - Quartiles and Percentiles

```python
import pandas as pd
import numpy as np

data = {'scores': [72, 68, 75, 80, 85, 90, 78, 82, 88, 92]}
df = pd.DataFrame(data)

print('Q1 (25%):', df['scores'].quantile(0.25))
print('Q2 (50% - Median):', df['scores'].quantile(0.50))
print('Q3 (75%):', df['scores'].quantile(0.75))
print('90th Percentile:', df['scores'].quantile(0.90))
print('IQR:', df['scores'].quantile(0.75) - df['scores'].quantile(0.25))
```
