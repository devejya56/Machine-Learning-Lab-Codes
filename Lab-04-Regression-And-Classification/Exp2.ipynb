# Lab 04: Exp2 - Multiple Linear Regression

**Objective:** Build and evaluate a multiple linear regression model with multiple features.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Create sample data with 2 features
X_train = np.array([[2, 1], [4, 2], [6, 3], [8, 4], [10, 5], [12, 6]])
y_train = np.array([5, 10, 15, 20, 25, 30])

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred_train = model.predict(X_train)

# Print model parameters
print('Model Coefficients:', model.coef_)
print('Model Intercept:', model.intercept_)
print()

# Calculate evaluation metrics
r2 = r2_score(y_train, y_pred_train)
mse = mean_squared_error(y_train, y_pred_train)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_train, y_pred_train)

print('=== Training Metrics ===' )
print('R2 Score:', r2)
print('MSE:', mse)
print('RMSE:', rmse)
print('MAE:', mae)
print()

# Test on new data
X_test = np.array([[3, 1.5], [5, 2.5], [7, 3.5]])
y_test_pred = model.predict(X_test)

print('=== Test Predictions ===' )
print('Input Features:', X_test)
print('Predictions:', y_test_pred)

# Feature importance (coefficients magnitude)
print()
print('=== Feature Importance ===' )
print('Feature 1 coefficient:', model.coef_[0])
print('Feature 2 coefficient:', model.coef_[1])
print('Higher coefficient = more important feature')
```
